```python
label_ext= ['EXT1', 'EXT2', 'EXT3', 'EXT4', 'EXT5', 'EXT6', 'EXT7', 'EXT8', 'EXT9', 'EXT10']
label_est = ['EST1', 'EST2', 'EST3', 'EST4', 'EST5', 'EST6', 'EST7', 'EST8', 'EST9', 'EST10']
label_agr = ['AGR1', 'AGR2', 'AGR3', 'AGR4', 'AGR5', 'AGR6', 'AGR7', 'AGR8', 'AGR9', 'AGR10']
label_cns = ['CSN1', 'CSN2', 'CSN3', 'CSN4', 'CSN5', 'CSN6', 'CSN7', 'CSN8', 'CSN9', 'CSN10']
label_opn = ['OPN1', 'OPN2', 'OPN3', 'OPN4', 'OPN5', 'OPN6', 'OPN7', 'OPN8', 'OPN9', 'OPN10']

label_ext_fa1 = ['EXT1', 'EXT2', 'EXT3', 'EXT4', 'EXT5', 'EXT6', 'EXT7', 'EXT8', 'EXT9', 'EXT10']
label_est_fa1 = ['EST1', 'EST2', 'EST3', 'EST5', 'EST6', 'EST7', 'EST8', 'EST9', 'EST10']
label_agr_fa1 = ['AGR1', 'AGR2', 'AGR3', 'AGR4', 'AGR5', 'AGR6', 'AGR7', 'AGR8', 'AGR9']
label_cns_fa1 = ['CSN1', 'CSN2', 'CSN4', 'CSN5', 'CSN6', 'CSN7', 'CSN8', 'CSN9']
label_opn_fa1 = ['OPN1', 'OPN3', 'OPN5', 'OPN7', 'OPN8',  'OPN10']

"""
itens_invertidos = ['EXT2','EXT8', 'EXT10','EXT4','EXT6',
                    'EST2',
                    'AGR1', 'AGR3', 'AGR5', 'AGR7',
                    'CSN2', 'CSN4', 'CSN6',  'CSN8', 
                    ]
"""

label_ext_fa = ['EXT2', 'EXT5', 'EXT7', 'EXT8', 'EXT9', 'EXT10']
label_est_fa = ['EST1', 'EST5', 'EST6', 'EST7', 'EST8', 'EST9']
label_agr_fa = ['AGR2', 'AGR4', 'AGR5', 'AGR6', 'AGR7', 'AGR9']
label_cns_fa = ['CSN2', 'CSN4', 'CSN5', 'CSN6', 'CSN7', 'CSN9']
label_opn_fa = ['OPN1', 'OPN3', 'OPN5', 'OPN7', 'OPN8', 'OPN10']

itens_ext = dados[label_ext_fa]
itens_est= dados[label_est_fa]
itens_agr = dados[label_agr_fa]
itens_cns = dados[label_cns_fa]
itens_opn = dados[label_opn_fa]
```

```python
def cronbach_alpha(df):
    # 1. Transform the df into a correlation matrix
    df_corr = df.corr()
    
    # 2.1 Calculate N
    # The number of variables equals the number of columns in the df
    N = df.shape[1]
    
    # 2.2 Calculate R
    # For this, we'll loop through the columns and append every
    # relevant correlation to an array calles "r_s". Then, we'll
    # calculate the mean of "r_s"
    rs = np.array([])
    for i, col in enumerate(df_corr.columns):
        sum_ = df_corr[col][i+1:].values
        rs = np.append(sum_, rs)
    mean_r = np.mean(rs)
    
   # 3. Use the formula to calculate Cronbach's Alpha 
    cronbach_alpha = (N * mean_r) / (1 + (N - 1) * mean_r)
    return cronbach_alpha
```

```python
# precisão direto

alpha = []
lista_de_questoes_por_fatores = [dados[label_ext_fa1], 
                     dados[label_est_fa1],
                     dados[label_agr_fa1],
                     dados[label_cns_fa1],
                     dados[label_opn_fa1],
]
for fator in lista_de_questoes_por_fatores:
  alpha.append(cronbach_alpha(fator))

print("Precisão Alpha de Cronbach para cada fator \n")
print(f"Alpha Extroversão: {round(alpha[0], 3): 10}")
print(f"Alpha estabilidade: {round(alpha[1], 3): 9}")
print(f"Alpha amabilidade: {round(alpha[2], 3):10}")
print(f"Alpha concienciosidade: {round(alpha[3], 3)}")
print(f"Alpha aberura: {round(alpha[4], 3): 14}")
```
